**Seaborn Demo**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy  as np
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
sns.set(style = 'dark')
```

**Categorical Plots**

* boxplot()

```python
sns.boxplot(x = 'day' , y = 'tip' , data = tips)
```

```python
sns.boxplot(x = 'day' , y = 'tip', data = tips, showfliers = False)
```

```python
sns.boxplot(x = 'day' , y = 'total_bill' , data = tips, hue = 'sex')
```

```python
sns.boxplot(x = 'day' , y = 'total_bill' , data = tips, hue = 'sex' , showfliers = False, showmeans = True, meanline = True)
```

```python
sns.boxplot(x = 'day' , y = 'total_bill' , data = tips, hue = 'smoker')
```

```python
# Add extra parameter related to matplotlib
#plt.figure(figsize = (15,5))
plt.gcf().set_size_inches(12,5)
sns.boxplot(x = 'day' , y = 'total_bill' , data = tips, hue = 'smoker', showfliers = False)
```

* countplot()

```python
sns.set(style = 'white')
```

```python
sns.countplot(x = 'day' , data = tips)
```

```python
sns.countplot(x = 'day' , data = tips , hue = 'sex')
```

```python
sns.countplot(x = 'smoker' , data = tips, hue = 'day' )
#plt.gcf().set_size_inches(15,5)
plt.figure(figsize = (15,5))
```

* barplot()

```python
sns.set(style= 'whitegrid')
```

```python
# estimator by default it is mean
sns.barplot(x = 'day' , y = 'total_bill' , data = tips, estimator = sum , ci = False)
```

```python
sns.barplot(x = 'sex' , y = 'total_bill' , data = tips, estimator = sum , ci = False , hue = 'time')
```

```python
sns.barplot(x = 'sex' , y = 'tip' , data = tips[tips['time'] == 'Dinner'] , hue = 'day' , estimator = sum)
plt.gcf().set_size_inches(15,5)
```

* catplot()

```python
sns.catplot(x ='day',y='total_bill',data=tips,hue = 'smoker',col = 'smoker' ,kind = 'bar' ,estimator=sum , ci = False)
```

```python
sns.set(style = 'dark')
sns.catplot(x ='size',y='total_bill',data=tips,col = 'day' ,kind = 'bar'  , ci = False)
plt.gcf().set_size_inches(15,5)
plt.suptitle('size vs. total_bill')
```

```python
sns.catplot(x ='size',y='total_bill',data=tips,col = 'day',row = 'smoker' ,kind = 'bar'  , ci = False)
plt.gcf().set_size_inches(10,5)
```

```python
selsize = tips[tips['size'].isin([2,3,4])]
sns.catplot(x = 'day' , y = 'total_bill' , data= selsize,col = 'size'  , kind = 'box' )
```

* striplot() & swarmplot()

```python
sns.stripplot(x = 'day' , y = 'total_bill' , data = tips)
```

```python
sns.stripplot(x = 'day' , y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.swarmplot(x = 'day' , y = 'total_bill', data = tips, hue = 'smoker' , dodge = True)
```

**Matrix Plots**

```python
ds = tips.pivot_table(index = 'day' , columns = 'time' , values = 'total_bill' , aggfunc = sum)
ds
```

```python
sns.heatmap(ds)
```

```python
sns.heatmap(ds , cmap = 'YlGnBu' ,annot = True, fmt = '.2f' )
```

```python
fl_summary  = flights.pivot('month' , 'year' , 'passengers')
sns.heatmap(fl_summary , annot = True , fmt = '.2f')
plt.gcf().set_size_inches(20,15)
```

```python
fl_summary  = flights.pivot('month' , 'year' , 'passengers')
sns.heatmap(fl_summary , annot = True , fmt = '.2f' , cbar = False)
plt.gcf().set_size_inches(20,15)
```

```python
cm = tips[['total_bill','tip','size']].corr()
print(cm)
sns.heatmap(cm, annot=True)
```
