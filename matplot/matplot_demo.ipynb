```python
tips = sns.load_dataset('tips')
```

```python
day_total = tips.groupby('day' , as_index = False)['total_bill'].sum()
day_total
```

```python
day_total['sex']
```

**Line Graph**

```python
plt.plot(day_total['day'], day_total['total_bill'])     #plot gives line graph(x-axis , y-axis)
plt.xlabel('Days')
plt.ylabel('Total_bill')
plt.title('Day vs. Total_bill')
```

```python
#plt.plot(day_total['day'], day_total['total_bill'], color = 'red' ,linestyle = 'dashed' , marker = 'D')
plt.plot(day_total['day'], day_total['total_bill'], 'bD--')
plt.xlabel('days')
plt.ylabel('bill')
plt.title('days vs. bill')
```

```python
male = tips[tips.sex == 'Male']
male_total = male.groupby('day', as_index = False)['total_bill'].sum()

female = tips[tips.sex == 'Female']
female_total = female.groupby('day', as_index = False)['total_bill'].sum()

plt.plot(male_total['day'] , male_total['total_bill'], color = 'red', label = 'Male')
plt.plot(female_total['day'] , female_total['total_bill'], color = 'blue', label = 'Female', linestyle = '--')
plt.legend()
plt.xlabel('days')
plt.ylabel('bill')
v = plt.title('days vs. bill')
```

**Text**

```python
plt.plot(day_total['day'], day_total['total_bill'])
days = day_total['day']
bills = day_total['total_bill']
for x,y in zip(days, bills):
    plt.text(x ,y + 80  , f"{y:.0f}" , fontdict = {'color' : 'green', 'size' : 15 ,'family' : 'calibri' } )
    print(x, f"{y:.2f}")
```

```python
countries = pd.read_json("https://restcountries.eu/rest/v2/all")
tc = countries.nlargest(5,'population')
plt.scatter(tc['area'] , tc['population'])
plt.xlabel('population')

a = tc['area']
p = tc['population']
for x,y in zip(a,p):
    plt.text(x , y, f"{y//10000000}-{x-'alpha3Code'} CR")
    print(x,y)
```

```python
countries = pd.read_json("https://restcountries.eu/rest/v2/all")
```

```python
asia = countries[countries['region'] == 'Asia'][['name', 'population','area']].sort_values(by = 'population', ascending = False).head(5)
```

```python
fig,(ax1,ax2 ) = plt.subplots(2,1)
fig.set_size_inches(15,15)
fig.suptitle('Asia VS. area & population')
ax1.bar(asia['name'] , asia['population'])
ax1.set_xticklabels(asia['name'], rotation = 'vertical')
ax2.bar(asia['name'] , asia['area'])
ax2.set_xticklabels(asia['name'], rotation = 'vertical')
```
